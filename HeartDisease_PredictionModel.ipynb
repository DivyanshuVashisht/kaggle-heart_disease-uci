### Importing the dataset file

```python
import pandas as pd
df = pd.read_csv('./heart.csv')
```

```python
df
```

```python
#Checking if there are any NaN values in the data set
df.isna().sum()
```

```python
#Slicing the target column and feature columns
y = df["target"]
x = df.drop("target", axis="columns")
```

```python
x
```

```python
y
```

### Specifying the Architecture

```python
from keras.layers import Dense
from keras.models import Sequential
from keras.utils import to_categorical
```

```python
y = to_categorical(y)
```

```python
y.ndim
```

```python
model=Sequential()
model.add(Dense(91, activation="relu", input_shape=(13,)))
model.add(Dense(2, activation="softmax"))
```

### Compiling the model

```python
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
```

### Training the model

```python
model.fit(x, y, epochs=15)
```
